**1. What is a parameter ?**

**2. What is correlation? What does negative correlation mean?**
  - Correlation is a statistical concept that shows the relationship between two variables. It tells us how one variable changes in relation to another variable. Correlation helps us understand whether the variables move in the same direction, in opposite directions, or if there is no relationship between them.

    - +1 → Perfect positive correlation

    - 0 → No correlation

  - **Negative correlation** means that when one variable increases, the other variable decreases. In simple words, both variables move in opposite directions.

**4. how does loss value help in determining whether the model is good or not?**
  - Loss value measures how well a machine learning model is performing. It shows the difference between the actual output and the predicted output. In simple words, loss tells us how much error the model is making.

    - Low loss value → Model predictions are close to actual values (Good model).

    - igh loss value → Model predictions are far from actual values (Poor model).

  - During training, the main goal is to minimize the loss value. As the model learns, the loss should decrease. If the loss keeps decreasing and becomes small, it means the model is improving.

**5. What are continuous and categorical variables ?**
  - Continuous variables are numerical variables that can take any value within a given range. They are measurable and can include decimal or fractional values.

    - For example: height, weight, temperature, and age.
    - These values can be measured precisely (like 170.5 cm, 65.8 kg).

  - Categorical Variables:

    - Categorical variables are variables that represent categories or groups. They do not represent numerical measurements but instead describe types or labels.

    - For example: gender (male/female), color (red/blue/green), city names, or pass/fail.

  - In simple words, continuous variables are measurable numbers, while categorical variables represent different groups or categories.

**6. How do we handle categorical variables in Machine Learning? What are the common techniques?**
  - Categorical variables cannot be directly used in most machine learning algorithms because models work with numerical data. So, we convert categorical data into numerical form using encoding techniques.

  - **Common Techniques to Handle Categorical Variables:**

    - Label Encoding – Assigns a unique number to each category
    
      (e.g., Red = 0, Blue = 1, Green = 2).

       It is simple and useful when categories have an order.

    - One-Hot Encoding – Creates separate binary (0 or 1) columns for each category. It is useful when categories do not have any order.

    - Ordinal Encoding – Used when categories have a meaningful order (e.g., Low = 1, Medium = 2, High = 3).

    - Target Encoding – Replaces categories with the mean value of the target variable. It is mostly used in advanced models.

**15. What is causation? Explain difference between correlation and causation with an example.**
  - Causation means that one variable directly causes a change in another variable. In other words, there is a cause-and-effect relationship between two variables.

  - **Difference between Correlation and Causation:**

    -  Correlation – It shows that two variables are related or move together, but it does not prove that one causes the other.

    -  Causation – It proves that one variable directly affects or causes a change in another variable.

  - Example:

    -  Ice cream sales and the number of drowning cases may show a positive correlation (both increase during summer). However, ice cream does not cause drowning. The real reason is hot weather (summer season).

     - Here, correlation exists, but there is no causation.

**16. What is an Optimizer? What are different types of optimizers? Explain each with an example.**
  - An optimizer is an algorithm used in machine learning and deep learning to minimize the loss function. Its main goal is to adjust the model’s parameters (weights and biases) so that the error between actual and predicted values becomes as small as possible.

  - In simple words, an optimizer helps the model learn better by reducing the loss during training.

  - **Different Types of Optimizers:**

    - Gradient Descent - It updates model parameters by moving in the direction of the negative gradient (minimum loss direction).
      - Example: Used in Linear Regression to minimize Mean Squared Error.

    - Stochastic Gradient Descent (SGD) - It updates parameters using one training example at a time, making it faster but slightly noisy.
      - Example: Commonly used in training neural networks.

    - Mini-Batch Gradient Descent - It updates parameters using a small batch of data instead of the full dataset.
      - Example: Mostly used in deep learning models for better efficiency.

    - Adam (Adaptive Moment Estimation) - It combines momentum and adaptive learning rates to improve performance. It is one of the most popular optimizers.
      - Example: Widely used in training deep learning models like CNNs and RNNs.

**21. What is feature scaling? How does it help in Machine Learning?**
  - Feature scaling is the process of transforming numerical features into a similar range or scale. It ensures that all input variables contribute equally to the model.

  - In many datasets, some features may have large values (like income in lakhs) while others have small values (like age). This difference can affect the model’s performance.

  - **It help in Machine Learning**

      1. Improves Model Performance – Algorithms like KNN, SVM, and Gradient Descent work better when features are on the same scale.

      2. Faster Convergence – It helps the model learn faster during training.

      3. Prevents Bias – Ensures that features with larger values do not dominate smaller ones.

      **Common Methods:**

      -  Standardization (StandardScaler)

      -  Normalization (MinMaxScaler)

**22. How do we perform scaling in Python?**
  - In Python, we perform feature scaling using the Scikit-learn library, mainly from the sklearn.preprocessing module.

  1. Standardization (StandardScaler)
It scales data so that the mean becomes 0 and standard deviation becomes 1.

          from sklearn.preprocessing import StandardScaler

          scaler = StandardScaler()
          X_train_scaled = scaler.fit_transform(X_train)
          X_test_scaled = scaler.transform(X_test)

  2. Normalization (MinMaxScaler)
  It scales data into a fixed range, usually between 0 and 1.

          from sklearn.preprocessing import MinMaxScaler

          scaler = MinMaxScaler()
          X_train_scaled = scaler.fit_transform(X_train)
          X_test_scaled = scaler.transform(X_test)

  - Important:

    -  Use fit_transform() on training data.

    -  Use only transform() on test data.
